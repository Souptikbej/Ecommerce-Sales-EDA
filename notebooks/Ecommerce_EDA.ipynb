## 6. Basic Outlier Detection

```python
df.boxplot(column=["Sales", "Profit"])
```

### Observation

- **Are there any extreme values?**
  - **Sales:** ✅ No significant outliers are visible. Sales values are fairly well distributed within the whiskers.
  - **Profit:** ⚠️ Yes. A few observations lie above the upper whisker (around **550–600**), indicating potential high-profit outliers. There is also a minimum profit of **-197.94**, but it falls within the expected range and is not shown as an extreme outlier.

- **Are they valid business cases or potential data errors?**
  - The high-profit values are **likely valid business cases** rather than data errors. They may represent:
    - High-value customer orders
    - Premium product sales
    - Large order quantities
    - Orders with little or no discount
  - There is **no evidence** from the boxplot that these values are data entry errors.
  - These records should be **investigated further** before deciding whether to remove them. In business datasets, such outliers often contain valuable insights.
